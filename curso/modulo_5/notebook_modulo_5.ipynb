```python
#%pip install -q google-genai

import google.genai as genai

API_KEY = "GEMINI_API_KEY"

client = genai.Client(api_key=API_KEY)

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="Explique overfitting em português"
)

print(response.text)
```

```python
prompt = "Quais são os principais riscos de se utilizar IA na cadeia de abastecimento?"
```

```python
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
)
print(response.text)
```

```python
def evaluate(response):
    print(response)
    
    factual = int(input("Factualidade (1-5): "))
    clarity = int(input("Clareza (1-5): "))
    consistency = int(input("Consistência (1-5): "))
    grounded = int(input("Base em evidência (1-5): "))
    
    return factual, clarity, consistency, grounded
```

```python
scores = evaluate(response)
print(scores)
```

```python
eval_prompt = f"""
Avalie a seguinte resposta quanto a correção factual e clareza:

{response}
"""
evaluation = client.models.generate_content(
                model="gemini-2.5-flash",
                contents=eval_prompt
            )
print(evaluation.text)
```

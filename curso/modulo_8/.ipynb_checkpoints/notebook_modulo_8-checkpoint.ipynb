```python
#%pip install -q google-genai

import google.genai as genai

API_KEY = "GEMINI_API_KEY"

client = genai.Client(api_key=API_KEY)

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="Explique overfitting em português"
)

print(response.text)
```

```python
def predict_demand():
    return {
        "forecast": 1200,
        "confidence": 0.78,
        "main_drivers": ["seasonality", "promotion"]
    }
```

```python
data = predict_demand()

prompt = f"""
Explique a seguinte previsão em termos simples:

Forecast: {data['forecast']}
Drivers: {data['main_drivers']}
"""

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
)
print(response.text)
```

```python
while True:
    question = input("Faça um questionamento sobre a previsão: ")
    if question == "exit":
        break
    
    prompt = f"""
Context:
{data}

User question:
{question}

Responda claramente:
"""
    response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
    )
    print(response.text)
```

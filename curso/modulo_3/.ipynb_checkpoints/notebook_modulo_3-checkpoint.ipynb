```python
#%pip install -q google-genai

import google.genai as genai

API_KEY = "GEMINI_API_KEY"

client = genai.Client(api_key=API_KEY)

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="Explique overfitting em português"
)

print(response.text)
```

```python
def ask(prompt):
    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt
    )
    return response.text
```

```python
task = "Como uma empresa pode reduzir os custos da cadeia de suprimentos sem prejudicar a qualidade do serviço?"
```

```python
zero = ask(task)
```

```python
few = ask("""
Q: Como reduzir os custos de inventário?
A: Otimizar níveis de estoque e fazer previsões.

Q: Como reduzir os custos de transporte?
A: Melhorar roteamento e consolidação.

Q: Como uma empresa pode reduzir os custos da cadeia de abastecimento?
A:
""")
```

```python
role = ask(f"""
Você é um especialista em cadeia de abastecimento.

Resposta:
{task}
""")
```

```python
cot = ask(f"""
Pense passo a passo.

Questão:
{task}
""")
```

```python
print("ZERO:\n", zero)
print("\nPOUCO:\n", few)
print("\nPAPEL:\n", role)
print("\nCOTA:\n", cot)
```
